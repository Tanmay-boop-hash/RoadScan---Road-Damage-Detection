```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os

dataset_path = '/content/drive/MyDrive/RDD_India'  # adjust if you named folder differently
print(os.listdir(dataset_path))
print("Train images:", len(os.listdir(f'{dataset_path}/train/images')))
print("Train labels:", len(os.listdir(f'{dataset_path}/train/labels')))
```

```python
yaml_content = """
path: /content/drive/MyDrive/RDD_India
train: train/images
val: val/images
test: test/images

nc: 5
names:
  0: longitudinal_crack
  1: transverse_crack
  2: alligator_crack
  3: pothole
  4: white_line_blur
"""

with open('/content/rdd_india.yaml', 'w') as f:
    f.write(yaml_content)

print("data.yaml created")
```

```python
from ultralytics import YOLO

# load pretrained YOLOv8 small model
model = YOLO('yolov8s.pt')

model.train(
    data='/content/rdd_india.yaml',
    epochs=30,
    imgsz=640,
    batch=16,
    name='rdd_india_v1',
    project='/content/drive/MyDrive/runs',   # saves directly to Drive
    patience=10,                             # stops early if no improvement
    device=0                                 # use GPU
)
```
